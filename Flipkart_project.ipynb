```python
#Calculate average ratings
avg_ratings=df['Ratings'].mean()
print(f'Average ratings:{avg_ratings}')
```

```python
#Plot 1
#Distribution of Ratings
plt.figure(figsize=(8,6))
sns.histplot(df['Ratings'],kde=True,bins=10)
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
#plot 2
#Bar plot of average discount by brand
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Discount',data=df,estimator=np.mean)
plt.title('Average Discount by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount(%)')
plt.xticks(rotation=45)
plt.show()
```

```python
#plot 3
#Count plot of number of products by brand
plt.figure(figsize=(10,6))
sns.countplot(x='Brand',data=df)
plt.title('Number of Products by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Average Ratings for Top 5 Brands by Count
top_brands=df['Brand'].value_counts().head(5).index
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Ratings',data=df[df['Brand'].isin(top_brands)],estimator=np.mean)
plt.title('Avearge Ratings for Top 5 Brands')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#plot 5
#Convert the MSP column to numeric(this will convert invalid parsing to NaN)
df['MSP']=pd.to_numeric(df['MSP'],errors='coerce')

#Now drop rows with NaN values in MSP column if needed
df=df.dropna(subset=['MSP'])

#Define price range segment
bins=[0,10000,20000,30000,50000,100000] #Define range
labels=['Low-end(<10k)','Budget(10k-20k)','Mid-range(20k-30k)','Premium(30k-50k)','Flagship(>50)']
```

```python
#Create a new column for price range
df['Price_Range']=pd.cut(df['MSP'],bins=bins,labels=labels)
#Count of mobiles in each price range
price_range_count=df['Price_Range'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=price_range_count.index,y=price_range_count.values)
plt.title('Mobile Count in Different Price Segment in India')
plt.xlabel('Price Range')
plt.ylabel('Number of Mobiles')
#Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
